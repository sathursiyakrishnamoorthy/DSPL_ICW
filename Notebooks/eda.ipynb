```python
import pandas as pd
import plotly.express as px
from pathlib import Path

# Load the clean data
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "Notebooks" else Path.cwd()
df = pd.read_csv("C://Users//User//Desktop//data science//Individual//DSPL_ICW//Proccessed//life_expectancy_clean.csv")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
df.head()
```

```python
# What does the data actually look like?
print("Life Expectancy:")
print(df['Life Expectancy'].describe())
print("\nGDP per Capita:")
print(df['GDP per Capita'].describe())
print(f"\nCountries: {df['Country Name'].nunique()}")
print(f"Regions: {df['Region'].unique()}")
print(f"Income groups: {df['IncomeGroup'].unique()}")
```

```python
# Chart 1: how has life expectancy changed by region over time?
region_avg = df.groupby(['Year', 'Region'])['Life Expectancy'].mean().reset_index()
fig = px.line(region_avg, x='Year', y='Life Expectancy', color='Region',
              markers=True, title='Avg Life Expectancy by Region (2013–2023)')
fig.show()
```

```python
# Chart 2: income group comparison
income_avg = df.groupby(['Year', 'IncomeGroup'])['Life Expectancy'].mean().reset_index()
fig = px.line(income_avg, x='Year', y='Life Expectancy', color='IncomeGroup',
              markers=True, title='Avg Life Expectancy by Income Group')
fig.show()
```

```python
# Chart 3: THE KEY INSIGHT - GDP vs Life Expectancy correlation
df_2023 = df[df['Year'] == 2023]
fig = px.scatter(df_2023, x='GDP per Capita', y='Life Expectancy',
                 color='Region', size='GDP per Capita',
                 hover_name='Country Name', log_x=True,
                 title='GDP per Capita vs Life Expectancy (2023)')
fig.show()
```

```python
# Chart 4: geographic distribution
df_2023 = df[df['Year'] == 2023]
fig = px.choropleth(df_2023, locations='Country Code',
                    color='Life Expectancy', hover_name='Country Name',
                    color_continuous_scale='Viridis',
                    title='Life Expectancy by Country (2023)')
fig.show()
```

```python
# Chart 5: which countries lead and lag?
df_2023 = df[df['Year'] == 2023].sort_values('Life Expectancy', ascending=False)
top_bottom = pd.concat([df_2023.head(10), df_2023.tail(10)])
fig = px.bar(top_bottom, x='Life Expectancy', y='Country Name',
             color='Region', orientation='h',
             title='Top 10 and Bottom 10 Countries by Life Expectancy (2023)')
fig.show()
```
